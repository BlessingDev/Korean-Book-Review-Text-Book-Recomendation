```python
import ipywidgets as widgets
from ipywidgets import interact, interactive, fixed, interact_manual
def df_interact(df, nrows=7, ncols=7):
    '''
    Outputs sliders that show rows and columns of df
    '''
    def peek(row=0, col=0):
        return df.iloc[row:row + nrows, col:col + ncols]

    row_arg = (0, len(df), nrows) if len(df) > nrows else fixed(0)
    col_arg = ((0, len(df.columns), ncols)
               if len(df.columns) > ncols else fixed(0))
    
    interact(peek, row=row_arg, col=col_arg)
    print('({} rows, {} columns) total'.format(df.shape[0], df.shape[1]))

def display_df(df, rows=pd.options.display.max_rows,
               cols=pd.options.display.max_columns):
    with pd.option_context('display.max_rows', rows,
                           'display.max_columns', cols):
        display(df)
```

### 유저별 분석

먼저 추천 필터링에서 가장 큰 의미를 갖는 각 유저 단위에서 분석해보려고 한다.

```python
review_grouped_by_user = rv1.groupby(["Author"])
```

```python
review_grouped_by_user[["Title"]].count()
```

총 11,530 명의 유저가 있는 것을 발견할 수 있었다.<br>
평균적으로 따지면 한 명의 유저당 두 권의 책을 리뷰했다는 것을 알 수 있다. 이 중 한 건의 리뷰만을 써서 유사도를 따질 수 없는 유저들의 리뷰는 삭제한다.

```python
user_group = review_grouped_by_user.groups
user_1_rv = []
for user in user_group.keys() :
    if len(user_group[user]) == 1 :
        user_1_rv.append(user)
len(user_1_rv)
```

유저 중에서 무려 7429명이 단 한 개의 리뷰만을 작성했다.

```python
rv1[~rv1["Author"].isin(user_1_rv)]
```

다음은 많은 리뷰를 단 유저들이 누구인지, 얼마나 많은 리뷰를 달았는지를 살펴보자.

```python
review_grouped_by_user[["Title"]].count().sort_values("Title", ascending=False).head(10)
```

다음은 리뷰 점수 분포를 살펴볼 것이다. 간단하게 하기 위해 디자인 점수와 내용점수를 합산할 것이다. 10점 만점이다.

```python
rv1 = rv1[~rv1["Author"].isin(user_1_rv)]
rv1["IntegratedRate"] = rv1["ConRate"] + rv1["DesRate"]
rv1["IntegratedRate"].iplot(kind="hist")
```

보다시피 10점이 대부분을 차지하고, 8점과 9점이 약간 있으며 그 이하는 거의 없다시피 하다.
